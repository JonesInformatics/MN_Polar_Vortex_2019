Tempatures in the Twin Cities Metro Area January-February 2019, see days 29-31 of the infamous "Polar Vortex."

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
file = "../data/poler_vortex_TC_2019.csv"

PolVor_df = pd.read_csv(file)

PolVor_df.head()
```

```python
# Set x axis to numerical value for month
x_axis_data = np.arange(1,60,1)
x_axis_data
```

```python
lows = PolVor_df["low"]
lows
```

```python
highs = PolVor_df["high"]
highs
```

```python
plt.plot(x_axis_data, lows)
```

```python
plt.xlabel("January-Feb 2019 Twin Cities Tempature")
plt.ylabel("Farenheit")

# giving a handle for lows
lows_handle, = plt.plot(x_axis_data, lows, label="lows")
# giving a handle for highs
highs_handle, = plt.plot(x_axis_data, highs, label="highs")
plt.legend(loc="lower right")
# Draw a horizontal line with 0.25 transparency to represent above/below zero marker
plt.hlines(0, 0, 60, alpha=0.40)
```
